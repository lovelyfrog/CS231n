#### 移动平均

验证移动平均在迭代一定次数之后与原始输入大小接近

```python
import numpy as np
```

```python
def moving_average(x, rho, running_mean=0):
    """
    x: [N,D]
    """
    mean = np.mean(x, axis=0)
    running_mean = rho * running_mean + (1 - rho) * mean
    print('mean:', mean)
    print('running_mean:', running_mean)
    return running_mean
```

```python
running_mean = 0
for t in range(100):
    x = 10 * np.random.randn(2000, 3) + 1
    rho = 0.9
    running_mean = moving_average(x, rho, running_mean)
```

可以发现上面在迭代了100次之后 移动平均与原始输入大小接近
